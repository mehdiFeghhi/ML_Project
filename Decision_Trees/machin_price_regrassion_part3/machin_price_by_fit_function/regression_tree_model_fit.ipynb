```python
def find_mean_column(column_data):

  item_not_null = 0
  calc_all = 0
  for item in column_data:
    if item != '-':
        calc_all += item
        item_not_null += 1
  
  return calc_all/item_not_null
```

```python
def find_replacing_for_null(X_train,list_of_categorical_column,list_of_continouse_column):
    dic_replacing_for_null = {} 
    for column in list_of_continouse_column:
        print(column)
        dic_replacing_for_null[column] = find_mean_column(X_train.loc[:, column])

    for column in list_of_categorical_column:
        dic_replacing_for_null[column] = X_train[column].mode()[0]


    return dic_replacing_for_null
```

```python
def find_data_set_range_for_each_coloumn_not_catogorical(data_set_feature,name_of_coloumn_not_categorical,number_of_cut):

    data_set_feature_details = data_set_feature.describe()
    diff = data_set_feature_details.iloc[7,:] - data_set_feature_details.iloc[3,:]
    diff = diff/number_of_cut
    print(diff)
    return diff
```

```python
def make_coloumn_not_catogorical_catogorical(name_coloumn_with_bais_range,data_set_feature, columns_name,number_of_cut):
    data_set_feature_copy = data_set_feature.copy()
    for column in columns_name:
      for index in data_set_feature.index:
            for c in range(number_of_cut):
                # print(data_set_feature[column])
                # print(data_set_feature.loc[j, column])
                # index(data_set_feature.loc[index,column])
                if data_set_feature.loc[index,column]!=None and (data_set_feature.loc[index,column] >= name_coloumn_with_bais_range[column]*c) and \
                   (data_set_feature.loc[index,column] < name_coloumn_with_bais_range[column]*(c+1)) and c != number_of_cut-1:

                    map = c
                
                elif c == number_of_cut-1 and data_set_feature.loc[index,column]!=None and (data_set_feature.loc[index,column] >= name_coloumn_with_bais_range[column]*c):

                     map = c
            
            if data_set_feature_copy[column][index] != None and map == number_of_cut - 1:
                  
                  data_set_feature_copy[column][index] = f'{column} >= {map*name_coloumn_with_bais_range[column]:.3f}'
            elif data_set_feature_copy[column][index] != None:

                  data_set_feature_copy[column][index] = f'{(map+1)*name_coloumn_with_bais_range[column]:.3f} > {column} >= {map*name_coloumn_with_bais_range[column]:.3f}'

    return data_set_feature_copy
```

```python
def make_node(dictionary_tree,father=None):
   
    if dictionary_tree['child'] == None:
        return Node_anytree(dictionary_tree['text'],parent=father)
    else:
        
        if father is None:
               root =  Node_anytree(dictionary_tree['text'])
        else:
               root = Node_anytree(dictionary_tree['text'],parent=father)
        
        for item in dictionary_tree['child']:
            make_node(item,father=root)
    
    return root
```

```python
class Node:
    def __init__(self,feature=None, feat_number=None,most_commen_child=None,choice_from_father=None, list_of_childs = None,
                 number_of_each_child_have_childs = None,variance=None,use_father_value=None,leaf=None,value=None):
      
        self.feature = feature
        self.most_commen_child = most_commen_child
        self.number_of_each_child_have_childs = number_of_each_child_have_childs
        self.feat_number = feat_number
        self.variance = variance
        self.use_father_value = use_father_value
        
        self.choice_from_father = choice_from_father
        self.childs = list_of_childs

        self.leaf = leaf
        self.value = value
    
    
    def is_leaf(self):
        return self.leaf    
    def show_node(self):
      enter = '\n'
      if self.is_leaf():
        return {"text":f"{self.choice_from_father} = {self._show_value()}","child":None}
      else :
        
        list_of_show_child =[]
        for child in self.childs:
            list_of_show_child.append(child.show_node())
        
        return {"text":f"{self.choice_from_father} _ {self.feature}","child": list_of_show_child}
    
    def _show_value(self):
        return self.value.show()

    def make_plot(self,name="Tree.png"):

        show_node_in_dic = self.show_node()
        
        root = make_node(show_node_in_dic)

        DotExporter(root).to_picture(name)
```

```python
class DecisionTree_Regression:
    def __init__(self,list_of_feature_name,calculate_function,list_of_continouse_column,max_depth=100,minmum_diference_variance=0,number_of_child_must_be_in_each_node=2):
        self.list_of_feature_name = list_of_feature_name
        self.max_depth = max_depth
        self.first = True
        self.how_to_calculate_value=calculate_function
        self.root = Node
        self.minmum_diference_variance = minmum_diference_variance
        self.list_of_continouse_column = list_of_continouse_column
        if len(list_of_continouse_column) < number_of_child_must_be_in_each_node: 
          self.number_of_child_must_be_in_each_node = number_of_child_must_be_in_each_node
        else :
          self.number_of_child_must_be_in_each_node = len(list_of_continouse_column)


    def _is_finished(self, depth,number_of_childs_have):
        if (depth >= self.max_depth
            or self.n_class_labels == 1
            or number_of_childs_have <= self.number_of_child_must_be_in_each_node):
            return True
        return False
    
    def _varicance(self, y):

        if len(y) == 0:
          return 0
        return np.var(y)

    def _create_split(self, X, list_of_category):

        dic_of_variable = {}
        # print(list_of_category)
        for item in list_of_category :
                idx = np.argwhere(X == item).flatten()
                dic_of_variable[item] = idx

        return dic_of_variable


    def _find_Remainder(self,dic_of_variable:dict,n,y):

      Remainder = 0
      # print(dic_of_variable)
      for key in  dic_of_variable:
        # print(key)
        Remainder += (len(dic_of_variable[key])/n) * self._varicance(y[dic_of_variable[key]])

      return Remainder

    def _information_gain(self, X, y, list_of_category):

        Varicance = self._varicance(y)
        dic_of_variable = self._create_split(X, list_of_category)
        n = len(y)
        Remainder = self._find_Remainder(dic_of_variable,n,y)
        return Varicance - Remainder

    def _best_split(self, X, y, features):
        best_score = -np.inf
        best_feat =  None, 
        best_list_of_category = None


        for feat in features:
            X_feat = X[:, feat]
            all_data_feat = self.data[:,feat]
            list_of_category = np.unique(all_data_feat)
            # for thresh in thresholds:
            score = self._information_gain(X_feat, y, list_of_category)
            if score > best_score:
                    best_score = score
                    best_feat = feat
                    best_list_of_category = list_of_category

        return best_feat, best_list_of_category,best_score
    
    def tree_show(self):
      
        return self.root.show_node()

    def make_plot(self,name):

        return self.root.make_plot(name)

    
    def _find_value(self,numpy_array_y,X_continus):

        return self.how_to_calculate_value(numpy_array_y,X_continus,self.list_of_continouse_column)
          
    def _build_tree(self, X, y,map_feature_number:list,father_value,
                    choice_from_father,X_continus,depth=0):

        self.n_samples, self.n_features = X.shape
        self.n_class_labels = len(np.unique(y))
        
        if self.first == True:
            self.data = X
            self.first = False
            


        # stopping criteria
        if self._is_finished(depth,len(X)) or len(map_feature_number)==0:
            if len(y) == 0 :
                return Node(choice_from_father=choice_from_father,
                                             value=father_value,
                                             variance=0,
                                             leaf=True,
                                             use_father_value = True,  
                                             )
            

            
            value = self._find_value(y,X_continus)

            return Node(choice_from_father=choice_from_father,
                                             value=value,
                                             variance=self._varicance(y),
                                             leaf=True,
                                             use_father_value = False,  
                                             )

        value = self._find_value(y,X_continus)

        best_feat, list_of_category,variance = self._best_split(X, y, map_feature_number)

        if variance <= self.minmum_diference_variance :
            return Node(choice_from_father=choice_from_father,
                                             value=value,
                                             variance=self._varicance(y),
                                             leaf=True,
                                             use_father_value = False,  
                                             )
        map_feature_number_copy = map_feature_number.copy()
        map_feature_number_copy.remove(best_feat)
        # grow children recursively
        # print(list_of_category)
        dic_of_variable = self._create_split(X[:, best_feat], list_of_category)
        
        list_of_childs = []
        number_of_attribut_each_child = []

        for key in dic_of_variable:
            
            most_commen_child = None
            number_item_most_commen = -1
            helper = 0
            number_of_attribut_each_child.append(len(dic_of_variable[key]))
            if len(dic_of_variable[key]) != 0 :

                  child = self._build_tree(X[dic_of_variable[key],:],
                                                    y[dic_of_variable[key]],
                                                    map_feature_number_copy,
                                                    value,
                                                    key,
                                                    X_continux[dic_of_variable[key],:],   
                                                    depth + 1)
                  list_of_childs.append(child)
                  
                  helper = len(X[dic_of_variable[key],:])
                  if helper > number_item_most_commen :

                      most_commen_child = child

            else:


                  list_of_childs.append(Node(choice_from_father=key,
                                             value=value,
                                             variance=0,
                                             leaf=True,
                                             use_father_value = True,  
                                             ))
              



        return Node(feature=self.list_of_feature_name[best_feat],
                    feat_number=best_feat,
                    most_commen_child=most_commen_child,
                    choice_from_father=choice_from_father,
                    list_of_childs=list_of_childs,
                    variance = self._varicance(y),
                    leaf=False,
                    value=value,
                    number_of_each_child_have_childs = number_of_attribut_each_child,
                    use_father_value = False
                    )
    
    def _traverse_tree(self, x,X_continux_item, node):
        if node.is_leaf():


            return node.value.value(X_continux_item)
        
        for child in node.childs:
          if x[node.feat_number] == child.choice_from_father:
            return self._traverse_tree(x,X_continux_item,child)

        return self._traverse_tree(x,X_continux_item,node.most_commen_child)


    def fit(self, X, y,X_continus,map_feature_number):
        map_feature_number = map_feature_number
        # father_most_commen_sample = np.argmax(np.bincount(y))
        father_value = self._find_value(y,X_continus)
        choice_from_father = "root"
        self.root = self._build_tree(X, y,map_feature_number,father_value,choice_from_father,X_continus)

    def predict(self, X,X_continus_item):
        predictions = [self._traverse_tree(X[Item_index],X_continus_item[Item_index], self.root) for Item_index in range(len(X))]
        return np.array(predictions)
```

```python
class Aproximation:

    def __init__(self,y,X,list_of_continouse_column):

        self.dic_of_continouse_use = {}
        self.fit_function = self._fit(y,X,list_of_continouse_column)
    def _fit(self,y,X,list_of_continouse_column):

        if len(X) > len(list_of_continouse_column) :

            rng = np.random.default_rng()

            size = len(list_of_continouse_column)
            idx = rng.choice(len(X), size=size,replace=False)
            for i in range(len(X)):
               size = len(list_of_continouse_column)
               idx = rng.choice(len(X), size=size,replace=False)
               X_selected = X[idx,:]

               Y_selected =y[idx]
               try :
                  res = np.linalg.solve(X_selected, Y_selected)
                  k = 0
                  for item in list_of_continouse_column:
                    self.dic_of_continouse_use[item] = k
                    k+=1
                  return res

               except :
                 
                      X_selected = self._add_noise(X_selected)
                      res = np.linalg.solve(X_selected, Y_selected)
                      k = 0
                      for item in list_of_continouse_column:
                          self.dic_of_continouse_use[item] = k
                          k+=1
                      return res


        elif len(X) == len(list_of_continouse_column) :
            k = 0
            for item in list_of_continouse_column:
                self.dic_of_continouse_use[item] = k
                k+=1
            try:
              return np.linalg.solve(X, y)
            except:
              x_selected = self._add_noise(X)
              return np.linalg.solve(x_selected, y)   

        elif len(X) < len(list_of_continouse_column):

                size = len(X)
                rng = np.random.default_rng()

                idx = rng.choice(len(list_of_continouse_column), size=size,replace=False)
                ix_sort = np.sort(idx)
                for item in ix_sort :
                     self.dic_of_continouse_use[list_of_continouse_column[item]] = item

                x_selected = X[:,ix_sort]
                try:
                  return np.linalg.solve(x_selected, y)   
                except:
                      x_selected = self._add_noise(x_selected)
                      return np.linalg.solve(x_selected, y)   


    def value(self,x):
        x_can_give = np.array([x[self.dic_of_continouse_use[key]] for key in self.dic_of_continouse_use])
        return np.dot(self.fit_function,x_can_give)

    def _add_noise(self,X):
        return X + np.random.randn(X.shape[0], X.shape[1])*0.001


    def show(self):

        string_return = ""
        el = 0
        for key in self.dic_of_continouse_use:
          string_return += key 
          string_return += " = "
          string_return += str(self.fit_function[el])
          string_return += "\n"
          el += 1
        
        return string_return
```
