```python
import pickle
import matplotlib.pyplot as plt

list_of_eq = []
list_of_eq_1 = []
list_of_rew = []
list_of_rew_1 = []
list_of_win = []
list_of_win_1 = []
with open('list_of_eq.pkl', 'rb') as f:
    list_of_eq = pickle.load(f)

with open('list_of_eq_1.pkl', 'rb') as f:
    list_of_eq_1 = pickle.load(f)

with open('list_of_rew.pkl', 'rb') as f:
    list_of_rew = pickle.load(f)

with open('list_of_rew_1.pkl', 'rb') as f:
    list_of_rew_1 = pickle.load(f)

with open('list_of_win.pkl', 'rb') as f:
    list_of_win = pickle.load(f)

with open('list_of_win_1.pkl', 'rb') as f:
    list_of_win_1 = pickle.load(f)


list_of_eq_all =  list_of_eq_1 + list_of_eq
list_of_rew_all = list_of_rew_1 + list_of_rew
list_of_win_all = list_of_win_1 + list_of_win



```

```python
def plot_lists(x, y, z):
    fig, ax = plt.subplots()
    ax.plot(range(len(x)), x, label='List equal')
    ax.plot(range(len(y)), y, label='List win')
    ax.plot(range(len(z)), z, label='List reward')
    ax.set_xlabel('Index')
    ax.set_ylabel('Values')
    ax.set_title('Three Lists Plot')
    ax.legend()
    plt.show()

# example usage
x_list = list_of_eq_all
y_list = list_of_win_all
z_list = list_of_rew_all
plot_lists(x_list, y_list, z_list)
```
